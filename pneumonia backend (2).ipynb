```python
IMAGE_SIZE=[224,224]
train_path=r"C:\Users\vaish\Downloads\archive (9)\chest_xray\train"
test_path=r"C:\Users\vaish\Downloads\archive (9)\chest_xray\test"
```

```python
from tensorflow.keras.applications.vgg16 import VGG16

# or (depending on your install)
# from keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16(input_shape=IMAGE_SIZE +[3],weights='imagenet',include_top=False)
```

```python
for layer in vgg.layers:
    layer.trainable=False
```

```python
from glob import glob
from tensorflow.keras.layers import Flatten, Dense, Dropout
```

```python
folders=glob(r"C:\Users\vaish\Downloads\archive (9)\chest_xray\train\*")
x=Flatten()(vgg.output)
```

```python
prediction=Dense(len(folders),activation='softmax')(x)
model=Model(inputs=vgg.input,outputs=prediction)
model.summary()
```

```python
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)

```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 horizontal_flip=True)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory(
    r'C:\Users\vaish\Downloads\archive (9)\chest_xray\train',
    target_size=(224,224),
    batch_size=10,
    class_mode='categorical'
)

```

```python
test_set = test_datagen.flow_from_directory(
    r'C:\Users\vaish\Downloads\archive (9)\chest_xray\test',
    target_size=(224,224),
    batch_size=10,
    class_mode='categorical'
)
```

```python
r = model.fit(
    training_set,
    validation_data=test_set,
    epochs=2,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
)
```

```python
x=image.img_to_array(img)
```

```python
img_data=preprocess_input(x)
```

```python
classes=model.predict(img_data)
```

```python
confidence = np.max(classes[0]) * 100
pred_class = np.argmax(classes[0])

if pred_class == 0:
    print(f"Result: Normal ({confidence:.2f}%)")
else:
    print(f"Result: Pneumonia ({confidence:.2f}%)")

```
